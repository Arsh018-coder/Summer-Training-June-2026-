# Step-4

# Questions & Answers

Q1-Does more social media usage lead to higher anxiety and stress?

```python
social_df['sm_bin'] = pd.cut(social_df.daily_social_media_hours,
                       bins=[0, 3, 5, 8],
                       labels=['Low (<3h)', 'Moderate (3–5h)', 'High (5h+)'])

q1 = social_df.groupby('sm_bin')[['stress_level', 'anxiety_level']].mean().round(2);
q1
```

```python
x = np.arange(3)
width = 0.35

plt.figure(figsize=(7, 5))
plt.title('Q1: Social Media Usage vs Stress & Anxiety')
plt.bar(x - width/2, q1['stress_level'],  width, label='Stress',  color='red', edgecolor='white')
plt.bar(x + width/2, q1['anxiety_level'], width, label='Anxiety', color='grey', edgecolor='white')
plt.xticks(x, q1.index)
plt.ylim(4, 7)
plt.xlabel('Daily Social Media Usage')
plt.ylabel('Average Score (1–10)')
plt.legend();
```

Q2-Does sleep affect academics and stress

```python
social_df['sleep_bin'] = pd.cut(social_df.sleep_hours,
                           bins=[3, 5, 6, 7, 8, 10],
                           labels=['<5h', '5–6h', '6–7h', '7–8h', '8h+'])

q2 = social_df.groupby('sleep_bin')[['academic_performance', 'stress_level']].mean().round(2)
q2
```

```python
x = q2[['academic_performance', 'stress_level']].plot(
    secondary_y='stress_level', kind='bar', color=['grey', 'red'], figsize=(7, 5)
)

plt.title('Q2: Sleep Hours vs Academic Performance & Stress')
x.set(xlabel='Sleep Duration', ylabel='GPA', ylim=(2.5, 3.5))
x.right_ax.set(ylabel='Stress Level (1–10)', ylim=(4.5, 6.5))
```

Q3-Are there differences in anxiety based on Gender?

```python
q3 = social_df.groupby('gender')[['stress_level', 'anxiety_level', 'addiction_level', 'sleep_hours']].mean().round(2)
q3
```

```python
q3.T.plot(kind='bar', figsize=(7, 5), color=['purple', 'lightblue'],
           edgecolor='white', width=0.6)
plt.title('Q3: Gender Differences in Mental Health Metrics')
plt.ylabel('Average Value')
plt.xlabel('Metric')
plt.xticks(rotation=30)
plt.legend(title='Gender')
```

Q4-Social Media Linked with different Mental health impact.

```python
grouped = social_df.groupby('platform_usage')[['stress_level', 'anxiety_level']].mean()

grouped.plot(kind='bar', figsize=(7, 5), color=['purple', 'orange'], edgecolor='white', width=0.7)

plt.xlabel('Social Media Platform')
plt.ylabel('Average Level (1-8)')
plt.title('Average Stress and Anxiety Levels by Social Media Platform')
plt.xticks(rotation=45, ha='right')
plt.ylim(0, 8)
plt.legend(['Stress Level', 'Anxiety Level']);
```

Q5-Does Social Media Affects Academic performance?

```python
academic_performance_by_sm = social_df.groupby('sm_bin')['academic_performance'].mean().round(2)

plt.figure(figsize=(7, 5))
plt.bar(academic_performance_by_sm.index, academic_performance_by_sm.values, color=['lightgray','black','crimson'])
plt.title('Average Academic Performance by Social Media Usage')
plt.xlabel('Social Media Usage')
plt.ylabel('Average Academic Performance');
```
