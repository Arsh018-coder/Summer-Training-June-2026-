```python
import pandas as pd
import numpy as np
```

```python
online = pd.read_csv('/content/online-retail.csv')
```

```python
online
```

```python
top_products=(
    online.groupby("Description")["Quantity"]
    .sum()
    .sort_values(ascending=False)
    .head(10)
)
print(top_products)
```

```python
months=[
"January",
"February",
"March",
"April",
"May",
"June",
"July",
"August",
"September",
"October",
"November",
"December"
]

online['InvoiceDate'] = pd.to_datetime(online['InvoiceDate'])
online['MonthName'] = online['InvoiceDate'].dt.month_name()
online['Revenue'] = online['Quantity'] * online['Price']

monthly_sales=(
online.groupby("MonthName")["Revenue"]
.sum()
.reindex(months)
)

print(monthly_sales);
```

```python
top_customers=(
    online.groupby("Customer ID")["Revenue"]
    .sum()
    .sort_values(ascending=False)
    .head(10)
)
print(top_customers)
```

```python
orders=(
online.groupby("Customer ID")["Invoice"]
.nunique()
.sort_values(ascending=False)
)
print(orders)
```

```python
basket=(
online.groupby("Invoice")["Revenue"]
.sum()
)
print(basket)
```

```python
country=(
online.groupby("Country")["Revenue"]
.sum()
.sort_values(ascending=False)
.head(10)
)
print(country)
```
